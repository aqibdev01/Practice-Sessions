```python
import seaborn as sns
import matplotlib as plt
```

```python
data = sns.load_dataset('penguins')
data
```

```python
sns.scatterplot(data=data,x='bill_length_mm',y='flipper_length_mm')
```

```python
sns.lineplot(data=data.head(100),x='bill_depth_mm',y='flipper_length_mm')
```

```python
sns.barplot(data=data,x='species',y='body_mass_g')
```

```python
sns.boxplot(data,x='sex',y='flipper_length_mm')
```

```python
sns.violinplot(data=data,x='island',y='body_mass_g')
```

```python
sns.histplot(data['body_mass_g'],bins=12,kde=True)
```

```python
sns.kdeplot(data['body_mass_g'],fill=True)
```

```python
sns.pairplot(data)
```

```python
corr = data[['bill_depth_mm','flipper_length_mm','body_mass_g']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
# plt.figure(figsize=(10,6))
sns.barplot(x= 'island',y='body_mass_g',data=data,estimator=sum)
```
